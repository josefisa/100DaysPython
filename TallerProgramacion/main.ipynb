```python
import matplotlib.pyplot as plt
```

```python
def recurrencia(a0,r,b,n):
    resultados = [a0]
    for _ in range(n):
        siguiente = resultados[-1] * r + b
        resultados.append(siguiente)
    return resultados

a0 = 438 #@param {"type":"slider","min":-10,"max":1000,"step":1}
r = 0.6 #@param {type:"slider", min:-2, max:2, step:0.1}
b = 2 #@param {type:"slider", min:-10, max:10, step:1}
n = 641 # @param {"type":"slider","min":1,"max":1000,"step":1}

resultados = recurrencia(a0, r, b, n)

plt.plot(range(n + 1), resultados)
plt.xlabel('n')
plt.ylabel('a(n)')
plt.title('Simulación de la Ecuación de Recurrencia')
plt.grid(True)
plt.show()
        
        
```
